```python
import csv
f=open('age.csv')
data=csv.reader(f)
result=[]
name=input('인구구조가 알고 싶은 지역의 이름')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))


plt.style.use('ggplot')
plt.title(name+' 지역의 인구구조')
plt.plot(result)   # 라인그래프
plt.show()
```

```python
import csv
f=open('gender.csv')
data=csv.reader(f)
m=[] #남성인구수의 (-)
f=[] #여성인구수
m2=[] #남성인구수
name=input('인구구조가 알고 싶은 지역의 이름: ')

for row in data:
  if name in row[0]:
    for i in row[3:104]:
      m.append(-int(i.replace(',','')))
      m2.append(int(i.replace(',','')))
    for i in row[106:]:
      f.append(int(i.replace(',','')))

import matplotlib.pyplot as plt
plt.title('포곡 지역의 남녀 인구 성별 분포')
plt.barh(range(101),m, label='남성')
plt.barh(range(101),f, label='여성')
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('궁금한 동네를 입력해주세요:')
for row in data :
  if name in row[0] :
    for i in range(3,104) :
      m.append(int(row[i]))
      f.append(int(row[i+103]))
    break 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.title(name+' 지역의 성별 인구 그래프')
plt.scatter(m,f, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
